```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/patrickmlong/Breast-Cancer-Wisconsin-Diagnostic-DataSet/master/data.csv")
```

```python
df.head(25)
```

```python
df.head()
```

```python
y = df['diagnosis']
```

```python
y.head()
```

```python
x = df.drop(['id','Unnamed: 32','diagnosis'], axis=1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y)
```

```python
le.classes_
```

```python
S = le.transform(y)
```

```python
y = S
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(xTrain,yTrain)
```

```python
knn.score(xTest,yTest)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(xTrain, yTrain)
```

```python
svclassifier.score(xTest,yTest)
```

```python
svcc = SVC(kernel='rbf')
svcc.fit(xTrain, yTrain)
```

```python
svcc.score(xTest,yTest)
```
