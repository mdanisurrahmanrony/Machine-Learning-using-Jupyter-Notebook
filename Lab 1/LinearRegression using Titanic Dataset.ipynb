```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv", sep='\t').dropna()
```

```python
df.head()
```

```python
y = df['Survived']
```

```python
x = df.drop(['PassengerId','Survived','Name','Ticket','Cabin','Embarked'], axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(x['Sex'])
```

```python
le.classes_
```

```python
S = le.transform(x['Sex'])
```

```python
x['Sex'] = S
```

```python
x
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
ln = LinearRegression()
ln.fit(xTrain,yTrain)
```

```python
ln.score(xTest,yTest)
```
